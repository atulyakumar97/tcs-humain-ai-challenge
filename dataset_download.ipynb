### Manually downloading a single image and cropping one face based on coordinates

```python
img = cv2.imread('image.jpeg') # reading downloaded image

ycorrect = img.shape[0]
xcorrect = img.shape[1]

# Coordinates for rectange
x1 = int(0.16622691292875902*xcorrect)
y1 = int(0.17150395778364103*ycorrect)
x2 = int(0.324538258575197*xcorrect)
y2 = int(0.325417766051011*ycorrect)

crop_img = img[y1:y2, x1:x2]

# Creating rectangle 
cv2.rectangle(img, (x1,y1),(x2,y2), (0, 255, 0), 2) 

plt.imshow(img[:,:,::-1])
```

## Automating face cropping and saving labels in a csv file in dataset folder

```python
import urllib.request
facecount=1

labels = pd.DataFrame(columns=['imagefile', 'emotion','age','ethnicity','gender'])

for i in range(len(df)):
    try:
        data = df.iloc[i,0]  
        f = open("downloaded_images/"+str(i+1)+".txt","w+")
        f.write(str(data))
        f.close()

        url = df.iloc[i,1]
        extension = "."+url.split('.')[-1]
        filename = str(i+1)+extension
        urllib.request.urlretrieve(url, "downloaded_images/"+filename)
        image = cv2.imread('downloaded_images/'+filename)

        for j in data:

            if j['label']==['Not_Face']:
                continue

            xcorrect = j['imageWidth']
            ycorrect = j['imageHeight']

            x1 = j['points'][0]['x']
            y1 = j['points'][0]['y']
            x2 = j['points'][1]['x']
            y2 = j['points'][1]['y']

            x1 = int(x1*xcorrect)
            y1 = int(y1*ycorrect)
            x2 = int(x2*xcorrect)
            y2 = int(y2*ycorrect)

            crop_face = image[y1:y2, x1:x2]
            plt.imshow(crop_img[:,:,::-1])

            face_filename = str(facecount)+extension
            save_location = "dataset/faceimages/"+face_filename
            cv2.imwrite(save_location,crop_face)

            emotion = j['label'][0]
            age = j['label'][1]
            ethnicity = j['label'][2]
            gender = j['label'][3]

            labels.loc[facecount-1] = [face_filename, emotion, age, ethnicity, gender]

            facecount=facecount+1    
        if i%10 == 0:
            print(i+1, 'image downloaded')
    except Exception as e:
        print('\nException '+str(e))
        print(i,j,'\n')
print(i+1, 'images downloaded')
print('Done')

labels.to_csv('dataset/labels.csv', index=False)
```
